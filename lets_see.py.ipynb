```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from ml_from_scratch.linear.classification import logistic

from typing import Optional, List

from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('ml_from_scratch/dataset/sn_ads.csv')
data
```

```python
data['Gender'] = data['Gender'].map({'Male': 1, 'Female': 0})
data
```

```python
data.info()
```

```python
# data = data.drop(columns=['Gender'])
data = data.drop(columns=['Gender'])
data
```

```python
X = data.iloc[:, [0, 1]].values
y = data.iloc[:, 2].values
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
x_tr, x_te, y_tr, y_te = train_test_split(X_scaled, y, test_size=0.20, random_state=42)
```

```python
clf = logistic.LogisticRegression(penalty = 'L1', max_iter=100000, learning_rate=0.001, alpha=0.01)
clf.fit(x_tr, y_tr)
```

```python
clf.coef_, clf.intercept_
```

```python
clf.coef_, clf.intercept_
```

```python
y_pred = clf.predict(x_te)
```

```python
print("Prediksi:", y_pred)
print("Label Aktual:", y_te)
```

```python
accuracy = np.mean(y_pred == y_te)
print(f"Akurasi: {accuracy * 100:.2f}%")
```
